```python
from selenium import webdriver
```

```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
# if chromedriver is not in your path, you’ll need to add it here
driver = webdriver.Chrome(r'C:\Users\Faiz\Downloads\chromedriver.exe')

driver.get('https://targetstudy.com/coaching/mba-entrance-exams-coaching-in-bareilly.htm')
```

```python
df=pd.DataFrame(columns=["Name","Address","phone_no"])
j=1
while j in (range(4)) :
    try:
        html=driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        outer_div = soup.find("div", attrs={'class' : "col-md-7 col-lg-7"})
        outer_div = list(outer_div)
        
        for i in range(len(outer_div)):
            try:
                a = outer_div[i].find_next("a").text
                data = list(outer_div[i].findAll('td'))
                b = data[1].text.strip() or data[0].text.strip() or data[-3].text.strip()
                c = data[-1].text
                
                df = df.append({'Name' : a , "Address" : b , "phone_no" : c}, ignore_index=True)
            except:
                
                pass
               
        next_link=driver.find_elements_by_link_text('»')
        WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.LINK_TEXT, '»'))).click()
        j+=1
    except:
    
        pass
```

```python
path="F:\\MBA_lucknow.csv"  
df.to_csv(path)
```
